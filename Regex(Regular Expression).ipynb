```python
import re
```

```python
text = "The phone number is important to speak with your family."
```

# ***Findall***

The findall() function returns a list containing all matches.

```python
res = re.findall("phone", text)
res
```

```python
res = re.findall("Hello", text)
res
```

# ***Sub***
The sub() function replaces the matches with the text of your choice

```python
text = "Cairo is the capital of Egypt and best Place I visited"
```

```python
res = re.sub("c", "7", text)
res
```

```python
# Number of replacements by specifying the count parameter
res = re.sub("c", "7", text, 2)
res
```

# ***Search***

The search() function is typically used to search for a specific pattern in a given string


```python
text = "The rain in Spain"
res = re.search("Egypt", text)
print(res)
```

```python
text = "Please contact us at support@example.com for help."
match_ = re.search(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,7}\b', text)
match_
```

# ***Split Function***

The split() function returns a list where the string has been split at each match

```python
text = "Hello my friend, I hope you are fine"
res = re.split(",", text)
res
```

```python
text = "Hello my friend, I hope you are fine"
res = re.split("/m",text, 2)
res
```

# ***More Examples***


```python
#Uppercase Letters
txt = "Hello World 1234"
x = re.findall(r"[A-Z]", txt)
x
```

```python
#LowerCase Letters
txt = "Hello World 1234"
x = re.findall(r"[a-z]", txt)
x
```

```python
#Digits
txt = "Hello World 1234"
x = re.findall(r"[0-9]", txt)
x
```

```python
#\w >>>> word character
txt = "Hello World 1234"
x = re.findall(r"\w", txt)
x
```

```python
#\W >>>> Non-word character
txt = "Hello World 1234 !"
x = re.findall(r"\W", txt)
x
```

```python
#\d >>>> digits
txt = "Hello World 1234"
x = re.findall(r"\d", txt)
x
```

```python
#\D >>>> Non-digits
txt = "Hello World 1234"
x = re.findall(r"\D", txt)
x
```

```python
# \s >>> space, tab, newlines
txt = "Hello World 1234"
x = re.findall(r"\s", txt)
x
```

```python
# \S >>> space, tab, newlines
txt = "Hello World 1234"
x = re.findall(r"\S", txt)
x
```

```python
pattern = r"ab*"
test_string = "abababababab"
res = re.findall(pattern, test_string)
res
```

```python
pattern = r"ab+"
test_string = "abb"
res = re.findall(pattern, test_string)
res
```

```python
pattern = r"ab{2,}"
test_string = "abbbbbb"
res = re.findall(pattern, test_string)
res
```

```python
phone_number_pattern = r"\(?\d{3}\)? \d{3}-\d{4}"
test_string = "My phone number is 123 456-7890"
res = re.findall(phone_number_pattern, test_string)
res
```

```python
#\b: Matches a word boundary (which is fine).
#Matches one or more word characters (letters, digits, and underscores), which is generally fine for matching the username part of an email.

txt = "Contact us at email@example.com or info@example.com"
pattern = r'\b\w+@[A-Za-z0-9.-]+\b'
res = re.findall(pattern, txt)
res
```
